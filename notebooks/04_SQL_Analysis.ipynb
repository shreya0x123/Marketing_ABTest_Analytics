```python
import pandas as pd
import sqlite3

# Load data
df = pd.read_csv('marketing_campaign.csv')  # adjust path if needed

# Create in-memory SQL DB
conn = sqlite3.connect(':memory:')
df.to_sql('marketing_data', conn, index=False, if_exists='replace')

print("Data loaded into SQL!")

# Run Query Example
query = """
SELECT campaign,
       SUM(clicked) * 1.0 / COUNT(*) AS CTR
FROM marketing_data
GROUP BY campaign;
"""
result = pd.read_sql(query, conn)
print(result)

conn.close()
```
