```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Load dataset (adjust path if needed)
df = pd.read_csv('marketing_campaign.csv')

# Create folders if not exists
os.makedirs('../visuals/plots', exist_ok=True)

# ======================
# 1. CTR Comparison
# ======================
plt.figure(figsize=(6, 4))
sns.barplot(x=df['campaign'], y=df['clicked'])
plt.title("Click-Through Rate (CTR) by Campaign")
plt.ylabel("CTR (Mean of Clicked)")
plt.savefig('../visuals/plots/ctr_comparison.png')
plt.show()

# ======================
# 2. Revenue Distribution
# ======================
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['campaign'], y=df['purchase_amount'])
plt.title("Revenue Distribution per Campaign")
plt.ylabel("Purchase Amount")
plt.savefig('../visuals/plots/revenue_boxplot.png')
plt.show()

# ======================
# 3. Combined Summary Plot
# ======================
fig, axs = plt.subplots(1, 2, figsize=(12, 4))

sns.barplot(x=df['campaign'], y=df['clicked'], ax=axs[0])
axs[0].set_title("Clicks by Campaign")
axs[0].set_ylabel("Avg Clicks")

sns.boxplot(x=df['campaign'], y=df['purchase_amount'], ax=axs[1])
axs[1].set_title("Revenue per Campaign")
axs[1].set_ylabel("Amount")

plt.tight_layout()
plt.savefig('../visuals/plots/summary_dashboard.png')
plt.show()

print("""
Dashboard Complete!  
Saved plots at: ../visuals/plots/
Files created:
- ctr_comparison.png
- revenue_boxplot.png
- summary_dashboard.png
""")
```
