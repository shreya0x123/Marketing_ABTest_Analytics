```python
import pandas as pd
from scipy.stats import ttest_ind

# Load the dataset
df = pd.read_csv('marketing_campaign.csv')  # adjust path if needed
print(df.head())

# Split revenue between A and B
rev_A = df[df['campaign'] == 'A']['purchase_amount']
rev_B = df[df['campaign'] == 'B']['purchase_amount']

# Perform t-test
t_stat, p_val = ttest_ind(rev_A, rev_B, equal_var=False)

print("t-test statistic:", t_stat)
print("p-value:", p_val)

print("Conclusion:", 
      "Campaign A leads to significantly higher revenue ðŸš€" if p_val < 0.05 
      else "No significant revenue difference.")
```
