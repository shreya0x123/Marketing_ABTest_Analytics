```python
import pandas as pd
from scipy.stats import chi2_contingency

# Load dataset
df = pd.read_csv('marketing_campaign.csv')  # change path if needed, e.g. '../data/...'

# Create contingency table for chi-square test
contingency_table = pd.crosstab(df['campaign'], df['clicked'])

# Run chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print("Chi-square Statistic:", chi2)
print("p-value:", p)
print("Degrees of freedom:", dof)
print("Expected Table:\n", expected)

# Conclusion
if p < 0.05:
    print("Conclusion: Campaign performance is significantly different â†’ one campaign is better!")
else:
    print("Conclusion: No significant difference between campaigns.")
```
